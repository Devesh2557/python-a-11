```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('ai_assistant_usage_student_life.csv')
```

```python
df.head(2)
```

```python
df.dtypes
```

```python
# df.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
lb = LabelEncoder()
```

```python
df['SessionID'] = lb.fit_transform(df['SessionID'])
df['StudentLevel'] = lb.fit_transform(df['StudentLevel'])
df['Discipline'] = lb.fit_transform(df['Discipline'])
df['SessionDate'] = lb.fit_transform(df['SessionDate'])
df['TaskType'] = lb.fit_transform(df['TaskType'])
df['FinalOutcome'] = lb.fit_transform(df['FinalOutcome'])
df['UsedAgain'] = lb.fit_transform(df['UsedAgain'])
```

```python
df.dtypes
```

```python
x = df.drop(columns = ['TaskType'])
y = df['TaskType']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

# LogisticRegression

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
```

```python
lr.fit(x_train, y_train)
```

```python
y_pred = lr.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test, y_pred)
```

# RandomForestClassifier

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier()
```

```python
rf.fit(x_train , y_train)
```

```python
y_pred = rf.predict(x_test)
```

```python
accuracy_score(y_test, y_pred)
```

# Naive Bayes

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
```

```python
x_train_new = sc.fit_transform(x_train)
```

```python
x_test_new = sc.transform(x_test)
```

```python
from sklearn.naive_bayes import GaussianNB
gb = GaussianNB()
```

```python
gb.fit(x_train_new, y_train)
```

```python
y_pred = gb.predict(x_test_new)
y_pred
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cn = confusion_matrix(y_test, y_pred)
cn
```

```python
accuracy_score(y_pred, y_test)
```
